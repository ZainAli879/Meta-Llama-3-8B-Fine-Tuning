```python
import json
import torch
from transformers import (AutoTokenizer,
                          AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          pipeline)
```

**Quantisation Configuration**

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          token=HF_TOKEN)

tokenizer.pad_token = tokenizer.eos_token
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config,
    token=HF_TOKEN
)
```

```python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
```

```python
def get_response(prompt):
  sequences = text_generator(prompt)
  gen_text = sequences[0]["generated_text"]
  return gen_text
```

```python
import streamlit as st
from transformers import LlamaTokenizer, LlamaForCausalLM, AutoTokenizer
import torch

# Load the model and tokenizer
model_name = "meta-llama/Meta-Llama-3-8B"
# use AutoTokenizer to automatically download the correct tokenizer type
tokenizer = AutoTokenizer.from_pretrained(model_name)
# Load the model
model = LlamaForCausalLM.from_pretrained(model_name)
# Function to generate a response from the LLaMA-3 model
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")

    # Ensure the model is in evaluation mode
    model.eval()

    # Generate a response (using a longer max_length if you want more context)
    with torch.no_grad():
        outputs = model.generate(inputs.input_ids, max_length=100, num_return_sequences=1, temperature=0.7)

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Streamlit app layout
st.title("LLaMA-3 Command Interface")
st.subheader("Enter a command, and LLaMA-3 will respond")

# User input field for the command
user_input = st.text_input("Enter your command:")

# Button to trigger response generation
if st.button("Get Response"):
    if user_input:
        # Generate and display the response from LLaMA-3
        response = generate_response(user_input)
        st.write("Response: ")
        st.write(response)
    else:
        st.warning("Please enter a command to get a response.")
```
